```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parâmetros do problema
L = 0.01  # Espessura da camada (m)
vw = -0.01  # Velocidade de sucção (m/s)
nu = 1.5e-5  # Viscosidade cinemática (m^2/s)
U = 10  # Velocidade do ar a 1 cm da placa (m/s)

# Parâmetros para a discretização
Ny = 11  # Número de pontos nodais
dy = L / (Ny - 1)  # Tamanho do passo na direção y
y_values = np.linspace(0, L, Ny)  # Valores de y de 0 a L

# Inicialização da matriz para armazenar os resultados
u_values = np.zeros(Ny)

# Condição inicial
u_values[0] = 0

# Discretização da equação diferencial usando diferenças finitas
for i in range(1, Ny):
    u_values[i] = u_values[i-1] + (dy * vw / nu) * (u_values[i-1] - U)

# Solução analítica
u_analytical = 1 - np.exp((y_values * vw) / nu)

# Plot dos resultados
plt.plot(y_values, u_values/U, label='Sol. Numérica', marker='o')
plt.plot(y_values, u_analytical, label='Sol. Analítica', linestyle='--')
plt.xlabel('Distância Vertical (m)')
plt.ylabel('u(y)/U')
#plt.title('Comparação da Solução Numérica e Analítica')
plt.title(f'Malha com {Ny} pontos nodais')
plt.legend()
plt.show()
```
